```python
#creating my baskets 

basket_napa=(sl[sl['region']=="Napa Valley"].groupby(['coustomer','wine'])['bool'].sum().unstack().reset_index().fillna(0).set_index('coustomer'))

basket_ruther=(sl[sl['region']=="Rutherford"].groupby(['day','wine'])['coustomer'].sum().unstack().reset_index().fillna(0).set_index('day'))


basket_oakvil=(sl[sl['region']=="Oakville"].groupby(['day','wine'])['coustomer'].sum().unstack().reset_index().fillna(0).set_index('day'))


basket_stags=(sl[sl['region']=="Stags Leap District"].groupby(['day','wine'])['coustomer'].sum().unstack().reset_index().fillna(0).set_index('day'))


basket_mount=(sl[sl['region']=="Mount Veeder"].groupby(['day','wine'])['coustomer'].sum().unstack().reset_index().fillna(0).set_index('day'))

```

```python
#crating a bool function for assotiation for the mechien 

def hot_encode(x):
    if(x<=5):
        return 0
    if(x>=5):
        return 1
```

```python
# running mechien learning alorithem in frame 


frequent_price=apriori(basket_napa, min_support=0.05, use_colnames=True)
rules=association_rules(frequent_price, metric="lift",min_threshold=1)
rules=rules.sort_values(['confidence','lift'],ascending=[False,False])
print(rules.head())


```
